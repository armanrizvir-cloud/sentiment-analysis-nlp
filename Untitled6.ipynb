```python
import pandas as pd

data = {
    "Restaurant Name": ["ABC Cafe", "Spice Hub", "Pizza Point", "Burger King", "Food Villa"],
    "City": ["Delhi", "Mumbai", "Delhi", "Lucknow", "Mumbai"],
    "Cuisines": ["North Indian", "Chinese", "Italian", "Fast Food", "North Indian"],
    "Price range": [2, 3, 2, 1, 4],
    "Has Online delivery": ["Yes", "No", "Yes", "No", "Yes"],
    "Aggregate rating": [4.2, 3.8, 4.5, 3.5, 4.7],
    "Votes": [120, 85, 200, 45, 300],
    "Longitude": [77.20, 72.87, 77.22, 80.94, 72.88], # Dummy Longitude values
    "Latitude": [28.61, 19.07, 28.63, 26.84, 19.09]   # Dummy Latitude values
}

df = pd.DataFrame(data)

df.to_csv("Dataset.csv", index=False)

df.head()
```

```python
df.info()
df.columns
```

```python
top_cuisines = df['Cuisines'].value_counts().head(3)

print(top_cuisines)
```

```python
percentage = (top_cuisines / len(df)) * 100

print(percentage)
```

```python
city_count = df['City'].value_counts()

print(city_count.head())
```

```python
avg_rating = df.groupby('City')['Aggregate rating'].mean()

print(avg_rating)
```

```python
print(avg_rating.idxmax())
print(avg_rating.max())
```

```python
print(df['Price range'].value_counts())
```

```python
import matplotlib.pyplot as plt

df['Price range'].value_counts().plot(kind='bar')

plt.title("Price Range Distribution")
plt.xlabel("Price Range")
plt.ylabel("Count")
plt.show()
```

```python
price_percent = (df['Price range'].value_counts()/len(df))*100

print(price_percent)
```

```python
online = (df['Has Online delivery']=="Yes").sum()

percentage = (online/len(df))*100

print(percentage)
```

```python
df.groupby('Has Online delivery')['Aggregate rating'].mean()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Aggregate rating'], bins=10)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
rating_range = pd.cut(df['Aggregate rating'],
                      bins=[0,1,2,3,4,5])

print(rating_range.value_counts())
```

```python
print("Average Votes:", df['Votes'].mean())
```

```python
top_combo = df['Cuisines'].value_counts().head(10)

print(top_combo)
```

```python
cuisine_rating = df.groupby('Cuisines')['Aggregate rating'].mean()

print(cuisine_rating.sort_values(ascending=False).head(10))
```
